```python
data = df.review.values.tolist()
print(len(data))
```

```python
from konlpy.tag import Twitter; t = Twitter()

tokens_ko = []
for word in data:
    tmp_list = t.morphs(str(word))
    tokens_ko.append(tmp_list)
    
result = []
tmp = []

for word in tokens_ko: 
    tmp = []
    for w in word:
        if w not in stop_words: 
            tmp.append(w)
    
    result.append(tmp)
        
print(len(result))     #195671
print(len(tokens_ko))
```

```python
#ko = nltk.Text(result, name='tokens_ko')  
ko_list = []
for nltk_word in result:
    ko = nltk.Text(nltk_word, name = 'tmp')
    ko_list.append(ko)
    
print(len(ko_list))
#print(len(set(ko.tokens))) # returns number of unique tokensb
```

```python
from konlpy.tag import Twitter; t = Twitter()

texts_ko_list = []
ko_list[0]

for ko in result:   #ko_list
    pos = lambda d: ['/'.join(p) for p in t.pos(d, stem=True, norm=True)]
    texts_ko = [pos(doc) for doc in ko]
    #for doc in ko:
     #   texts_ko = t.pos(doc, stem = True, norm = True)
    texts_ko_list.append(texts_ko)

print(len(texts_ko_list))
```

```python
noun_adj_list = []
texts_ko_list_cp = texts_ko_list

for texts_ko in texts_ko_list_cp:
    tmp_list = []
    for word in texts_ko:
        try:
            #print(str(word).split('/')[1].split('\'')[0])
            _ = str(word).split('/')[1].split('\'')[0]
            if _ == 'Adjective' or _ == 'Noun' or _ == 'Verb':    #형용사, 명사
                tmp_list.append(word)
        except:
            print(word)
    
    noun_adj_list.append(tmp_list)

print((noun_adj_list[0]))
```

```python
texts_ko_list_except = []
texts_ko_example = []

for ele in noun_adj_list:
    tmp_texts_ko = []
    for ele_word in ele:
        _ = str(ele_word).split('/')[0].split('\'')[1]
        if _ not in stop_adj:
            tmp_texts_ko = tmp_texts_ko + ele_word
            texts_ko_example.append(ele_word)
    texts_ko_list_except.append(tmp_texts_ko)
```

```python
# 유클리드 거리

from scipy.spatial import distance
dst = np.zeros([1019,1019])

for i in range(1019):
    for j in range(1019):
        _dst = np.sqrt(np.sum(np.square(data_arr[i] - data_arr[j])))
        _dst /= (np.sqrt(np.sum(np.square(data_arr[i]))) + np.sqrt(np.sum(np.square(data_arr[j]))))
        dst[i,j] = _dst
dst
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit=10, start=2, step=1):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = gensim.models.wrappers.LdaMallet(mallet_path, corpus=tf_ko, num_topics=7, id2word=dictionary_ko)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts_ko_list_except, dictionary=dictionary_ko, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary_ko, corpus=tf_ko, texts=texts_ko_list_except, start=3, limit=10, step=1)
```

```python
# Show graph
limit=10; start=3; step=1;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()


# Print the coherence scores
for m, cv in zip(x, coherence_values):
    print("Num Topics =", m, " has Coherence Value of", round(cv, 4))
```
