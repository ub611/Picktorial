```python
dir_list = ['TOP_QUALITY_GAME', 'TOP_SELLIG_GAME', 'TOP_SELLING_PAID']
csv_list = []
for d in dir_list :
    for file in os.listdir(d + "/"):
        for csv in os.listdir(d + "/" + file):
            if csv.endswith(".csv") :
                csv_list.append(d + "/" + file + "/" + csv)
            else :
                print(file)
```

```python
from hanspell import spell_checker
from pandas import DataFrame
from konlpy.tag import Mecab
import pandas as pd
import os
import sys
from konlpy.tag import Kkma
import pickle
from operator import eq

#긍정 데이터 추출
def pos_review(_data):
    s_reviews = list(_data[_data.columns[0]])
    s_reviews.append(_data.columns[0])
    r_reviews = list(_data[_data.columns[2]])
    
    reviews = []
    for i in range(len(s_reviews)):
        if(eq(s_reviews[i],5)):
            reviews.append(r_reviews[i])
        else:
            print('')
    
    return reviews


#hanspell 맞춤법 검사
def do_hanspell(reviews):
    new_reviews = []
    for i in range(len(reviews)):
        try:
            result = spell_checker.check(reviews[i])
            new_reviews.append(result.as_dict()["checked"])
        except:
            print('')
    return new_reviews
```

```python
def make_file(_data, name):
    reviews = pos_review(_data)
    new_reviews = do_hanspell(reviews)
    dat = {'name' : name, 'review' : new_reviews}
    df = pd.DataFrame(data=dat)
    df.to_csv('Reviews/' + name + '.csv', sep=',', encoding='utf-8')
```
