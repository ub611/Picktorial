```python
from hanspell import spell_checker
from pandas import DataFrame
from konlpy.tag import Mecab
import pandas as pd
import os
import sys
from konlpy.tag import Kkma
import pickle
from operator import eq
```

```python
file = 'Five Nights at Freddy\'s_review' + '.csv'
#[x for x in os.listdir('/Users/minhyeonlee/Desktop/Crawling/data/') if x[-4:]=='.csv']
_data = pd.read_csv(file)
```

```python
#긍정 데이터 추출
def pos_review(_data):
    s_reviews = list(_data[_data.columns[0]])
    s_reviews.append(_data.columns[0])
    r_reviews = list(_data[_data.columns[2]])
    
    reviews = []
    for i in range(len(s_reviews)):
        if(eq(s_reviews[i],5)):
            reviews.append(r_reviews[i])
    else:
        print('')
    
    return reviews
```

```python
#hanspell 맞춤법 검사
def do_hanspell(reviews):
    new_reviews = []
    for i in range(len(reviews)):
        try:
            result = spell_checker.check(reviews[i])
            new_reviews.append(result.as_dict()["checked"])
        except:
            print(i)
    return new_reviews
```

```python
reviews = pos_review(_data)
#print(reviews)
```

```python
new_reviews = do_hanspell(reviews)
#print(new_reviews)
```

```python
f = open(file + ".txt", 'w')
review_string = str(new_reviews)

review_string = review_string.replace("[", "")
review_string = review_string.replace("]", "")
review_string = review_string.replace("'", "")
review_string = review_string.replace(",", "")

f.write(review_string)
```

```python
f.close()
```
