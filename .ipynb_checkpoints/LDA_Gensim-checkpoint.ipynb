```python
r = open('example_word.txt', mode = 'rt', encoding='utf-8')
doc_word = r.read()

doc_word = doc_word.replace("[", "")
doc_word = doc_word.replace("]", "")
doc_word = doc_word.replace("'", "")
doc_word = doc_word.split(",")
#doc_word = list(set(doc_word))

doc_matrix = [[" " for _ in range(len(doc_word))] for _ in range(2)]
doc_range = list(map(str, range(len(doc_word))))

for i in range(len(doc_word)):
    doc_matrix[0][i] = doc_range[i]
    doc_matrix[1][i] = doc_word[i]
    
```

```python
print(word_matrix)
print(doc_matrix)
```

```python
# Create Dictionary

id2word = corpora.Dictionary(word_matrix)
# Create Corpus
texts = doc_matrix
 
# Term Document Frequency=
corpus = [id2word.doc2bow(text) for text in texts]
 
# View
print(corpus)
id2word[0]
```
